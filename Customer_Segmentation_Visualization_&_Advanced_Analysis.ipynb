```python
# Load libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
from google.colab import files
uploaded = files.upload()

df = pd.read_csv(list(uploaded.keys())[0])

# Convert 'TotalCharges' to numeric, coercing errors to NaN
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
# Drop rows where 'TotalCharges' became NaN (if any)
df.dropna(subset=['TotalCharges'], inplace=True);

# Drop non-numeric identifier column 'customerID'
df = df.drop('customerID', axis=1)

# Basic info
print(df.head())
print(df.info())
print(df.describe())

# Missing values
print(df.isnull().sum())

# Correlation - now only on numeric columns
plt.figure(figsize=(12,6))
sns.heatmap(df.select_dtypes(include=np.number).corr(), annot=True, cmap='viridis')
plt.show()

# Histograms
df.hist(figsize=(12,8))
plt.show()
```

TASK:- 2

```python
# Task 2: Data Cleaning
import pandas as pd

# (Use your existing dataframe 'df' from Task 1)

print("Before Cleaning:")
print("Missing values:\n", df.isnull().sum())
print("\nDuplicate rows:", df.duplicated().sum())
```

```python
# Step 1: Standardize column names
df.columns = df.columns.str.lower().str.strip().str.replace(" ", "_")

print("\nColumn names standardized successfully!")
print(df.columns)
```

```python
# Step 2: Handle Missing Values
# Choose a strategy that fits your dataset:
# Example strategy:
# Numerical columns â†’ fill with mean
# Categorical columns â†’ fill with mode

num_cols = df.select_dtypes(include=['int64', 'float64']).columns
cat_cols = df.select_dtypes(include=['object']).columns

# Fill numeric missing values with mean
for col in num_cols:
    df[col] = df[col].fillna(df[col].mean())

# Fill categorical missing values with mode
for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

print("\nMissing values handled!")
```

```python
# Step 3: Remove duplicated rows
df.drop_duplicates(inplace=True)
print("\nDuplicates removed!")
# Final Summary
print("\nAfter Cleaning:")
print("Missing values:\n", df.isnull().sum())
print("\nDuplicate rows:", df.duplicated().sum())

print("\nDataset is cleaned and ready for analysis!")
```

```python
# Task 4: Customer Segmentation Visualization

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
# (Use cleaned dataframe df)
# Step 1: Create Tenure Segmentation
bins = [0, 12, 36, 100]  # tenure groups
labels = ["0-12 months", "13-36 months", "37+ months"]

df["tenure_group"] = pd.cut(df["tenure"], bins=bins, labels=labels, include_lowest=True)

print("Tenure groups created successfully!")
print(df["tenure_group"].value_counts())
```

```python
# Step 2: Pie Chart (or Donut Chart)

plt.figure(figsize=(8, 8))
sizes = df["tenure_group"].value_counts()
colors = sns.color_palette("pastel")

plt.pie(sizes, labels=sizes.index, autopct="%1.1f%%", startangle=90, colors=colors)
centre_circle = plt.Circle((0, 0), 0.70, fc='white')  # makes donut chart
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title("Customer Distribution by Tenure Group (Donut Chart)", fontsize=14)
plt.show()
```

```python
# Step 3: Clustered Bar Chart (Matplotlib)

avg_monthly_charges = df.groupby("tenure_group")["monthlycharges"].mean()

plt.figure(figsize=(8, 5))
bars = plt.bar(avg_monthly_charges.index, avg_monthly_charges.values)

plt.title("Average Monthly Charges Across Tenure Segments")
plt.xlabel("Tenure Group")
plt.ylabel("Average Monthly Charges")

# Add annotations (values on bars)
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 1, round(yval, 2),
             ha='center', va='bottom')

plt.show()

```

```python
# Step 4: Interactive Visualization with Plotly

fig = px.bar(
    avg_monthly_charges.reset_index(),
    x="tenure_group",
    y="monthlycharges",
    title="Interactive Clustered Bar Chart â€“ Average Monthly Charges",
    text="monthlycharges"
)

fig.update_traces(texttemplate='%{text:.2f}', textposition='outside')
fig.update_layout(yaxis_title="Average Monthly Charges",
                  xaxis_title="Tenure Group")
fig.show()
```

```python
# Task 5: ADVANCED ANALYSIS

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# (Use the cleaned dataframe: df)

# PART 1: Group customers by tenure group and compute stats

# Convert 'churn' column to numerical representation (0 for No, 1 for Yes) for aggregation
df['churn'] = df['churn'].map({'No': 0, 'Yes': 1})

tenure_stats = df.groupby("tenure_group").agg({
    "monthlycharges": ["mean", "median", "max", "min"],
    "totalcharges": ["mean", "median", "max", "min"],
    "churn": "mean"   # churn rate per group (now numeric)
})

print("ðŸ“Œ Tenure-based Statistics:")
display(tenure_stats)


# Bar chart: churn rate by tenure group
plt.figure(figsize=(7, 5))
sns.barplot(x=df["tenure_group"], y=df["churn"])
plt.title("Churn Rate by Tenure Group")
plt.ylabel("Churn Rate")
plt.show()
```

```python
# PART 2: Demographic analysis (Gender, SeniorCitizen)

# Churn rate by Gender
plt.figure(figsize=(6, 4))
sns.barplot(x="gender", y=df["churn"].astype(int), data=df)
plt.title("Churn Rate by Gender")
plt.show()

# Churn rate by Senior Citizen
plt.figure(figsize=(6, 4))
sns.barplot(x="seniorcitizen", y=df["churn"].astype(int), data=df)
plt.title("Churn Rate by Senior Citizen Status")
plt.show()
```

```python
# Payment Method
plt.figure(figsize=(10, 5))
sns.barplot(x="paymentmethod", y=df["churn"].astype(int), data=df)
plt.xticks(rotation=45)
plt.title("Churn Rate by Payment Method")
plt.show()

# Contract Type
plt.figure(figsize=(7, 5))
sns.barplot(x="contract", y=df["churn"].astype(int), data=df)
plt.title("Churn Rate by Contract Type")
plt.show()
```

```python
# PART 4: Trend Visualization (Monthly Charges vs Tenure)
# ---------------------------------------------------------

plt.figure(figsize=(10, 5))
sns.lineplot(x="tenure", y="monthlycharges", data=df)
plt.title("Monthly Charges Trend Across Customer Tenure")
plt.xlabel("Tenure (Months)")
plt.ylabel("Monthly Charges")
plt.show()
```

```python
# PART 5: Churn vs Lifecycle Stage Visualization

plt.figure(figsize=(10, 6))
sns.boxplot(x="tenure_group", y="totalcharges", hue="churn", data=df)
plt.title("Total Charges Across Lifecycle Stages (Churn vs Non-Churn)")
plt.show()
```
